```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import mstats

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, ConfusionMatrixDisplay

# -----------------------------------------------------------------------------
# 1) Load & preprocess data
# -----------------------------------------------------------------------------
def load_and_preprocess(path):
    df = pd.read_parquet(path)
    # Ensure timestamp column is datetime64
    df["timestamp"] = pd.to_datetime(df["timestamp"])
    # Sort by coin and time
    df.sort_values(["coin_id", "timestamp"], inplace=True)
    # Create target: 1 if next close > current close, else 0
    df["target_direction"] = (
        df.groupby("coin_id", observed=True)["close"]
          .transform(lambda x: (x.shift(-1) > x).astype(int))
    )
    # Drop rows without a valid target
    df.dropna(subset=["target_direction"], inplace=True)
    return df

# -----------------------------------------------------------------------------
# 2) Feature engineering
# -----------------------------------------------------------------------------
def add_features(df):
    df = df.copy()
    # Sort by coin and time
    df.sort_values(["coin_id", "timestamp"], inplace=True)

    # Volume ratio feature
    df["volume_ratio"] = df["taker_buy_quote_asset_volume"] / df["quote_asset_volume"]
    
    # Log returns and range features
    df["log_ret"]   = df.groupby("coin_id")["close"].transform(lambda x: np.log(x).diff())
    df["range_pct"] = (df["high"] - df["low"]) / df["open"]
    df["oc_pct"]    = (df["close"] - df["open"]) / df["open"]

    # Simple moving average and exponential moving average
    df["sma_10"] = df.groupby("coin_id")["close"].transform(lambda x: x.rolling(10).mean())
    df["ema_10"] = df.groupby("coin_id")["close"].transform(lambda x: x.ewm(span=10, adjust=False).mean())

    # Rolling volatility and momentum over multiple windows
    windows = [5, 10, 20, 50, 100]
    for w in windows:
        df[f"ret_ma_{w}"] = df.groupby("coin_id")["log_ret"].transform(lambda x: x.rolling(w).mean())
        df[f"vol_{w}"]    = df.groupby("coin_id")["log_ret"].transform(lambda x: x.rolling(w).std())
    df["mom_diff_5_50"] = df["ret_ma_5"] - df["ret_ma_50"]

    # Trend ratio between two SMAs
    df["sma_7"]  = df.groupby("coin_id")["close"].transform(lambda x: x.rolling(7).mean())
    df["sma_30"] = df.groupby("coin_id")["close"].transform(lambda x: x.rolling(30).mean())
    df["trend_ratio"] = df["sma_7"] / df["sma_30"]

    # Fourier features for daily and weekly seasonality
    sec = df["timestamp"].values.astype("int64") // 10**9
    day, week = 24*60*60, 7*24*60*60
    df["sin_day"]  = np.sin(2 * np.pi * sec / day)
    df["cos_day"]  = np.cos(2 * np.pi * sec / day)
    df["sin_week"] = np.sin(2 * np.pi * sec / week)
    df["cos_week"] = np.cos(2 * np.pi * sec / week)

    return df

# -----------------------------------------------------------------------------
# 3) Load and feature-engineer all intervals (no scaling yet)
# -----------------------------------------------------------------------------
intervals = ["10min", "1h", "4h", "1d"]
train_feats, test_feats = {}, {}

for iv in intervals:
    print(f"--- Interval: {iv} ---")
    tr = load_and_preprocess(f"data/resampled/train_{iv}.parquet")
    te = load_and_preprocess(f"data/resampled/test_{iv}.parquet")
    
    tr_feat = add_features(tr).dropna(subset=["target_direction"]).reset_index(drop=True)
    te_feat = add_features(te).dropna(subset=["target_direction"]).reset_index(drop=True)
    
    train_feats[iv] = tr_feat
    test_feats[iv]  = te_feat
    print(f"  train shape: {tr_feat.shape}, test shape: {te_feat.shape}")

# -----------------------------------------------------------------------------
# 4) Modeling: classification & evaluation with pipeline (scaling inside CV)
# -----------------------------------------------------------------------------
feature_cols = [
    c for c in train_feats["10min"].columns
    if c not in ("coin_id", "timestamp", "target_direction")
]

tscv       = TimeSeriesSplit(n_splits=5)
param_grid = {"clf__C": [0.1, 1.0, 10.0]}

for iv in intervals:
    print(f"\n=== Evaluation @ {iv} ===")
    tr_df = train_feats[iv].dropna(subset=feature_cols + ["target_direction"])
    te_df = test_feats[iv].dropna(subset=feature_cols + ["target_direction"])
    
    for coin in tr_df["coin_id"].unique():
        print(f"\n-- Coin: {coin} --")
        tr_coin = tr_df[tr_df["coin_id"] == coin]
        te_coin = te_df[te_df["coin_id"] == coin]
        if tr_coin.empty or te_coin.empty:
            print("  No data, skipping")
            continue
        
        X_train, y_train = tr_coin[feature_cols], tr_coin["target_direction"]
        X_test,  y_test  = te_coin[feature_cols], te_coin["target_direction"]
        
        pipe = Pipeline([
            ("scaler", StandardScaler()),
            ("clf",    LogisticRegression(
                class_weight="balanced", solver="liblinear", max_iter=1000
            ))
        ])
        grid = GridSearchCV(
            pipe, param_grid=param_grid,
            cv=tscv, scoring="roc_auc", n_jobs=-1
        )
        grid.fit(X_train, y_train)
        best = grid.best_estimator_
        print("  Best C:", grid.best_params_["clf__C"])
        
        y_pred  = best.predict(X_test)
        y_proba = best.predict_proba(X_test)[:, 1]
        print(classification_report(y_test, y_pred, zero_division=0))
        print("  Test ROC AUC:", roc_auc_score(y_test, y_proba))
        
        disp = ConfusionMatrixDisplay.from_predictions(
            y_test, y_pred, normalize="true"
        )
        disp.ax_.set_title(f"{iv} — {coin} Confusion Matrix")
        plt.tight_layout()
        plt.show()

# -----------------------------------------------------------------------------
# 5) Backtesting: retrain on full train set, then simulate strategy
# -----------------------------------------------------------------------------
periods_per_year = {"10min": 365*24*6, "1h": 365*24, "4h": 365*6, "1d": 365}
backtests = {iv: {} for iv in intervals}

print("\n=== P&L Backtest ===")
for iv in intervals:
    print(f"\n--- Interval: {iv} ---")
    for coin in train_feats[iv]["coin_id"].unique():
        tr_coin = (train_feats[iv][train_feats[iv]["coin_id"] == coin]
                   .dropna(subset=feature_cols + ["target_direction"]))
        te_coin = (test_feats[iv][test_feats[iv]["coin_id"] == coin]
                   .dropna(subset=feature_cols + ["target_direction"]))
        if tr_coin.empty or te_coin.empty:
            print(f"  {coin}: no data, skipping")
            continue

        X_tr, y_tr = tr_coin[feature_cols], tr_coin["target_direction"]
        pipe = Pipeline([
            ("scaler", StandardScaler()),
            ("clf",    LogisticRegression(
                class_weight="balanced", solver="liblinear", max_iter=1000
            ))
        ])
        grid = GridSearchCV(pipe, param_grid=param_grid,
                            cv=tscv, scoring="roc_auc", n_jobs=-1)
        grid.fit(X_tr, y_tr)
        best = grid.best_estimator_

        df_bt = te_coin.sort_values("timestamp").reset_index(drop=True)
        # Generate trading signal as a pandas Series, then shift
        signal = pd.Series(best.predict(df_bt[feature_cols]), index=df_bt.index)
        df_bt["signal"] = signal.shift(1).fillna(0).astype(int)
        # Compute next period return
        df_bt["bar_ret"] = df_bt["close"].pct_change().shift(-1)
        df_bt.dropna(subset=["bar_ret"], inplace=True)
        # Filter out extreme jumps
        df_bt = df_bt[df_bt["bar_ret"].abs() <= 0.20]

        # Compute strategy and buy-and-hold returns
        df_bt["strat_ret"]   = df_bt["signal"] * df_bt["bar_ret"]
        df_bt["strat_ret_w"] = mstats.winsorize(df_bt["strat_ret"], limits=[0.005, 0.005])
        df_bt["cum_strat"]   = (1 + df_bt["strat_ret_w"]).cumprod()
        df_bt["cum_bh"]      = (1 + df_bt["bar_ret"]).cumprod()

        plt.figure(figsize=(10, 4))
        plt.plot(df_bt["timestamp"], df_bt["cum_strat"], label="Strategy")
        plt.plot(df_bt["timestamp"], df_bt["cum_bh"],    label="Buy & Hold")
        plt.title(f"{iv} | {coin} Equity Curve")
        plt.xlabel("Time")
        plt.ylabel("Cumulative Return")
        plt.legend()
        plt.tight_layout()
        plt.show()

        tot_strat = df_bt["cum_strat"].iloc[-1] - 1
        tot_bh    = df_bt["cum_bh"].iloc[-1]    - 1
        sharpe    = df_bt["strat_ret_w"].mean() / df_bt["strat_ret_w"].std() \
                    * np.sqrt(periods_per_year[iv])
        print(f"  {coin}: Strategy = {tot_strat:.2%}, B&H = {tot_bh:.2%}, Sharpe = {sharpe:.2f}")

        backtests[iv][coin] = df_bt[["timestamp", "strat_ret_w", "bar_ret"]].copy()

# -----------------------------------------------------------------------------
# 6) Equal-weight portfolio backtest
# -----------------------------------------------------------------------------
print("\n=== Equal-weight Portfolio Backtest ===")
for iv in intervals:
    print(f"\n--- Interval: {iv} ---")
    pieces = []
    for coin, df_bt in backtests[iv].items():
        tmp = df_bt.set_index("timestamp").rename(
            columns={"strat_ret_w": f"{coin}_strat", "bar_ret": f"{coin}_bh"}
        )
        pieces.append(tmp)
    if not pieces:
        print("  no backtests for this interval.")
        continue

    port = pd.concat(pieces, axis=1).fillna(0)
    strat_cols = [c for c in port.columns if c.endswith("_strat")]  
    bh_cols    = [c for c in port.columns if c.endswith("_bh")]    

    port["ew_strat_ret"] = port[strat_cols].mean(axis=1)
    port["ew_bh_ret"]    = port[bh_cols].mean(axis=1)
    port["cum_ew_strat"] = (1 + port["ew_strat_ret"]).cumprod()
    port["cum_ew_bh"]    = (1 + port["ew_bh_ret"]).cumprod()

    plt.figure(figsize=(10, 4))
    plt.plot(port.index, port["cum_ew_strat"], label="Strategy (EW)")
    plt.plot(port.index, port["cum_ew_bh"],    label="Buy & Hold (EW)")
    plt.title(f"{iv} — EW Portfolio Equity Curve")
    plt.xlabel("Time")
    plt.ylabel("Cumulative Return")
    plt.legend()
    plt.tight_layout()
    plt.show()

    tot_strat = port["cum_ew_strat"].iloc[-1] - 1
    tot_bh    = port["cum_ew_bh"].iloc[-1]    - 1
    sharpe    = port["ew_strat_ret"].mean() / port["ew_strat_ret"].std() \
                * np.sqrt(periods_per_year[iv])
    print(f"  EW Portfolio: Strategy = {tot_strat:.2%}, B&H = {tot_bh:.2%}, Sharpe = {sharpe:.2f}")
```
